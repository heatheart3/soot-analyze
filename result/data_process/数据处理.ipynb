```python
f = open("E:/Projects/Java/analysis_soot/if_throwPatternResult3.0.txt", "r")
l = f.read()
print(l)
```

```python
d = {}
c = []
sourceCode = []
```

```python
d = {}
c = []
m = []
scl = []
l1 = l.split("\n")
for i in l1:
    if i.__contains__("Class:"):
        ll = i.replace("Class:", "").replace("Method:", "").replace("Source Code Line:", "").replace("   ", "").split(" ")
        
        c.append(ll[0])
        m.append(ll[1])
        scl.append(ll[2])
    
d["Class"] = c
d["Method"] = m
d["SourceCodeLine"] = scl
```

```python
df = pd.DataFrame(d)
df.to_csv("if_throwPattern3.0.csv",index=False)
```
